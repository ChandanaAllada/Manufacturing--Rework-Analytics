```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Sample Manufacturing Data
data = {
    'Line': ['Line-A', 'Line-A', 'Line-B', 'Line-B', 'Line-C', 'Line-C'],
    'Defect_Type': ['Soldering', 'Assembly', 'Soldering', 'Testing', 'Assembly', 'Testing'],
    'Rework_Count': [45, 30, 60, 20, 35, 25]
}
df = pd.DataFrame(data)

# 2. Analysis: Which Line has most rework
rework_by_line = df.groupby('Line')['Rework_Count'].sum().sort_values(ascending=False)
print("Total Rework by Production Line:")
print(rework_by_line)

# 3. Analysis: Which Defect is most common
rework_by_defect = df.groupby('Defect_Type')['Rework_Count'].sum().sort_values(ascending=False)
print("\nTotal Rework by Defect Type:")
print(rework_by_defect)

# 4. Visualization - Bar Graph
plt.figure(figsize=(8,5))
rework_by_line.plot(kind='bar', color='skyblue')
plt.title('Rework Count by Production Line')
plt.ylabel('Number of Reworks')
plt.xlabel('Production Line')
plt.show()
```
